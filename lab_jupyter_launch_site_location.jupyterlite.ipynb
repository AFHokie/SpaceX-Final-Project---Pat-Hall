```python
launch_sites_df['marker_color'] = ''

# Iterate over each launch site in the launch_sites_df DataFrame
for index, site in launch_sites_df.iterrows():
    # Get the success and failure counts for each site
    success_count = len(spacex_df[(spacex_df['Launch Site'] == site['Launch Site']) & (spacex_df['class'] == 1)])
    failure_count = len(spacex_df[(spacex_df['Launch Site'] == site['Launch Site']) & (spacex_df['class'] == 0)])

    # Determine the color based on the success rate
    if success_count > failure_count:
        color = 'green'
    else:
        color = 'red'

    # Update the 'marker_color' column for the current site
    launch_sites_df.at[index, 'marker_color'] = color

# Display the updated launch_sites_df DataFrame
launch_sites_df


```

```python
# Add marker_cluster to current site_map
# Iterate over each row in the spacex_df DataFrame
# Add a new column 'marker_color' to the spacex_df DataFrame
spacex_df['marker_color'] = ''

# Iterate over each launch record in the spacex_df DataFrame
for index, launch in spacex_df.iterrows():
    # Determine the color based on the launch outcome
    if launch['class'] == 1:
        color = 'green'  # Success (1)
    else:
        color = 'red'  # Failure (0)
    
    # Update the 'marker_color' column for the current launch record
    spacex_df.at[index, 'marker_color'] = color

# Display the updated spacex_df DataFrame
spacex_df

```

```python
# Iterate over each launch record in the spacex_df DataFrame
# Create a map centered around the mean of the coordinates
map_center = [launch_sites_df['Lat'].mean(), launch_sites_df['Long'].mean()]
site_map = folium.Map(location=map_center, zoom_start=4)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Create a dictionary to store the number of results by launch site
results_by_site = {}

# Iterate over each launch record in the spacex_df DataFrame
for index, record in spacex_df.iterrows():
    # Determine the color based on the launch outcome
    color = 'green' if record['class'] == 1 else 'red'
    
    # Add a marker to the marker_cluster for each launch record with a color based on the outcome
    marker = folium.Marker([record['Lat'], record['Long']], icon=folium.Icon(color=color))
    marker_cluster.add_child(marker)
    
    # Update the results_by_site dictionary
    if record['Launch Site'] in results_by_site:
        results_by_site[record['Launch Site']] += 1
    else:
        results_by_site[record['Launch Site']] = 1

# Add yellow circles with labels indicating the number of results at each launch site
for site, count in results_by_site.items():
    circle = folium.CircleMarker(
        location=launch_sites_df.loc[launch_sites_df['Launch Site'] == site, ['Lat', 'Long']].values[0],
        radius=10,
        color='yellow',
        fill=True,
        fill_color='yellow',
        popup=f"{site}: {count} results"
    )
    site_map.add_child(circle)

# Display the map
site_map

```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

from geopy.distance import geodesic

# Define the coordinates of the launch sites
launch_sites = {
    'CCAFS LC-40': (28.562302, -80.577356),
    'CCAFS SLC-40': (28.563197, -80.576820),
    'KSC LC-39A': (28.573255, -80.646895),
    'VAFB SLC-4E': (34.632834, -120.610745)
}

# Calculate the distances from each launch site to its proximities
for site, coords in launch_sites.items():
    for proximity, proximity_coords in launch_sites.items():
        if site != proximity:
            distance = geodesic(coords, proximity_coords).kilometers
            print(f"Distance from {site} to {proximity}: {distance:.2f} kilometers")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
